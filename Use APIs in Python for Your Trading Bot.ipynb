### Step 1: Import Libraries

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2: Define Your API Key and URL (Replace with your actual key)

Signup to get your FREE API key: https://www.alphavantage.co/support/#api-key
Keep it safe and do not share it with anyone. 

```python
api_key = 'REPLACE_WITH_YOUR_ALPHA_VANTAGE_API_KEY'
symbol = 'AAPL'
url = f'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={symbol}&apikey={api_key}&outputsize=compact'
```

### Step 3: Request the Data

```python
response = requests.get(url)
data = response.json()
```

### Step 4: Parse JSON to DataFrame

```python
daily_data = data['Time Series (Daily)']
df = pd.DataFrame(daily_data).T 
df.index = pd.to_datetime(df.index)
df = df.astype(float)
df = df.sort_index()
```

### Step 5: Plot the Closing Prices

```python
plt.figure(figsize=(10,5))
plt.plot(df['4. close'], label=f'{symbol} Closing Price')
plt.title(f'{symbol} Stock Price (API Data)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid()
plt.show()
```
